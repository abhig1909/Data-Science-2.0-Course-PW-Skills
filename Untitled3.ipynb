```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Class:
A class is a blueprint or template for creating objects. It defines attributes (variables) and methods (functions) that the objects will have.

Object:
An object is an instance of a class. It represents a real-world entity and has its own state and behavior defined by the class.

Example:

python
Copy
Edit
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"Car: {self.brand} {self.model}")

# Creating objects (instances of the class)
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

car1.display_info()  # Output: Car: Toyota Corolla
car2.display_info()  # Output: Car: Honda Civic
```

```python
Q2. Name the four pillars of OOPs.
The four pillars of Object-Oriented Programming (OOP) are:

Encapsulation – Wrapping data and methods into a single unit (class) to restrict direct access.
Abstraction – Hiding complex implementation details and exposing only essential features.
Inheritance – Acquiring properties and behaviors of a parent class in a child class.
Polymorphism – The ability of an object to take multiple forms (e.g., method overriding and method overloading).
```

```python
Q3. Explain why the __init__() function is used. Give a suitable example.
The __init__() function is a constructor method in Python that is automatically called when an object is created. It initializes the object's attributes.

Example:

python
Copy
Edit
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_details(self):
        print(f"Student Name: {self.name}, Age: {self.age}")

# Creating an object
s1 = Student("Alice", 20)
s1.show_details()  # Output: Student Name: Alice, Age: 20
```

```python
Q4. Why self is used in OOPs?
In Python, self represents the instance of a class and is used to access variables and methods within the class. It differentiates instance variables from local variables inside methods.

Example:

python
Copy
Edit
class Animal:
    def __init__(self, species):
        self.species = species  # 'self.species' is an instance variable

    def show_species(self):
        print(f"This is a {self.species}")

# Creating an object
a1 = Animal("Dog")
a1.show_species()  # Output: This is a Dog
```

```python
Q5. What is inheritance? Give an example for each type of inheritance.
Inheritance allows a class (child class) to inherit attributes and methods from another class (parent class).

Types of Inheritance:
Single Inheritance – A child class inherits from a single parent class.

python
Copy
Edit
class Animal:
    def sound(self):
        print("Animals make sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()  # Inherited method
d.bark()
Multiple Inheritance – A child class inherits from multiple parent classes.

python
Copy
Edit
class Parent1:
    def func1(self):
        print("This is Parent1")

class Parent2:
    def func2(self):
        print("This is Parent2")

class Child(Parent1, Parent2):
    pass

obj = Child()
obj.func1()
obj.func2()
Multilevel Inheritance – A child class inherits from another child class.

python
Copy
Edit
class Grandparent:
    def grandparent_func(self):
        print("This is Grandparent")

class Parent(Grandparent):
    def parent_func(self):
        print("This is Parent")

class Child(Parent):
    def child_func(self):
        print("This is Child")

c = Child()
c.grandparent_func()
c.parent_func()
c.child_func()
Hierarchical Inheritance – Multiple child classes inherit from a single parent class.

python
Copy
Edit
class Vehicle:
    def general_info(self):
        print("All vehicles have wheels")

class Car(Vehicle):
    def car_info(self):
        print("Cars have four wheels")

class Bike(Vehicle):
    def bike_info(self):
        print("Bikes have two wheels")

car_obj = Car()
bike_obj = Bike()

car_obj.general_info()
car_obj.car_info()

bike_obj.general_info()
bike_obj.bike_info()
Hybrid Inheritance – A combination of different types of inheritance.

python
Copy
Edit
class A:
    def funcA(self):
        print("This is A")

class B(A):
    def funcB(self):
        print("This is B")

class C(A):
    def funcC(self):
        print("This is C")

class D(B, C):
    def funcD(self):
        print("This is D")

obj = D()
obj.funcA()  # Inherited from A
obj.funcB()  # Inherited from B
obj.funcC()  # Inherited from C
obj.funcD()  # Own method
```
