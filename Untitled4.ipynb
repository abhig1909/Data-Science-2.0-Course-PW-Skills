```python
Q1. Create a Vehicle class with an __init__ method having instance variables name_of_vehicle, max_speed, and average_of_vehicle.
python
Copy
Edit
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name = name_of_vehicle
        self.max_speed = max_speed
        self.average = average_of_vehicle

    def display_info(self):
        print(f"Vehicle: {self.name}, Max Speed: {self.max_speed} km/h, Average: {self.average} km/l")

# Example usage
v1 = Vehicle("Bike", 120, 40)
v1.display_info()
```

```python
Q2. Create a child class Car from the Vehicle class which will inherit from it. Create a method named seating_capacity which takes capacity as an argument and returns the name of the vehicle and its seating capacity.
python
Copy
Edit
class Car(Vehicle):  # Inheriting from Vehicle
    def seating_capacity(self, capacity):
        return f"{self.name} has a seating capacity of {capacity} people."

# Example usage
car1 = Car("Sedan", 200, 15)
print(car1.seating_capacity(5))
```

```python
Q3. What is multiple inheritance? Write a Python code to demonstrate multiple inheritance.
Multiple inheritance is when a class inherits from more than one parent class.

python
Copy
Edit
class Parent1:
    def func1(self):
        print("Function from Parent1")

class Parent2:
    def func2(self):
        print("Function from Parent2")

class Child(Parent1, Parent2):  # Inheriting from both Parent1 and Parent2
    def func3(self):
        print("Function from Child")

# Example usage
obj = Child()
obj.func1()  # Inherited from Parent1
obj.func2()  # Inherited from Parent2
obj.func3()  # Defined in Child
```

```python
Q4. What are getter and setter in Python? Create a class and create a getter and a setter method in this class.
Getters and setters are used to access and modify private attributes in a class.

python
Copy
Edit
class Student:
    def __init__(self, name, age):
        self._name = name  # Private attribute
        self._age = age    # Private attribute

    # Getter method
    def get_age(self):
        return self._age

    # Setter method
    def set_age(self, new_age):
        if new_age > 0:
            self._age = new_age
        else:
            print("Age must be positive!")

# Example usage
s = Student("Alice", 20)
print(s.get_age())  # Output: 20

s.set_age(25)  # Updating age
print(s.get_age())  # Output: 25

s.set_age(-5)  # Invalid update
```

```python
Q5. What is method overriding in Python? Write a Python code to demonstrate method overriding.
Method overriding occurs when a child class redefines a method from the parent class with different behavior.

python
Copy
Edit
class Animal:
    def sound(self):
        print("Animals make different sounds")

class Dog(Animal):
    def sound(self):  # Overriding the parent class method
        print("Dog barks")

# Example usage
a = Animal()
a.sound()  # Output: Animals make different sounds

d = Dog()
d.sound()  # Output: Dog barks
```
